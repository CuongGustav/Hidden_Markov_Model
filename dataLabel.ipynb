### Convert Sample Frequency to 16kHz
- Mục đích nhằm chuyển các tệp âm thanh về chung một tần số mẫu, nhằm mục đích đảm bảo tính đồng nhất cho dữ liệu đầu vào của mô hình, giảm thiểu sai lệch giữa các tệp âm thanh có tần số mẫu khác nhau.

```python
import os
import librosa
import soundfile as sf

# Đường dẫn tới thư mục dữ liệu gốc và thư mục đích
input_folder = 'train'
output_folder = 'trainNew'
target_sample_rate = 16000  # Tần số mẫu mong muốn (ví dụ: 16 kHz)

# Tạo thư mục đích nếu chưa tồn tại
os.makedirs(output_folder, exist_ok=True)

# Duyệt qua tất cả các folder con trong thư mục gốc
for subfolder in os.listdir(input_folder):
    subfolder_path = os.path.join(input_folder, subfolder)
    
    # Kiểm tra xem đó có phải là một thư mục hay không
    if os.path.isdir(subfolder_path):
        # Tạo thư mục con tương ứng trong thư mục đích
        output_subfolder_path = os.path.join(output_folder, subfolder)
        os.makedirs(output_subfolder_path, exist_ok=True)
        
        # Duyệt qua tất cả các file âm thanh trong thư mục con
        for filename in os.listdir(subfolder_path):
            if filename.endswith('.wav'):  # Chỉ xử lý các file .wav
                input_path = os.path.join(subfolder_path, filename)
                output_path = os.path.join(output_subfolder_path, filename)

                # Đọc tệp âm thanh và thay đổi tần số mẫu
                audio, sr = librosa.load(input_path, sr=target_sample_rate)
                
                # Lưu tệp âm thanh với tần số mẫu mới vào thư mục đích
                sf.write(output_path, audio, target_sample_rate)
                print(f'Chuyển đổi tần số mẫu cho {filename} trong {subfolder} thành công.')

print("Chuyển đổi tần số mẫu hoàn tất cho tất cả các tệp!")
```

### Gộp file có cùng nội dung vào chung nhãn

```python
import os
import shutil

# Thư mục chứa các folder A, B, ..., T
source_dir = "trainNew"
# Thư mục đích để lưu dữ liệu tổ chức lại
destination_dir = "datasets"

# Danh sách tên thư mục mới thay vì 01, 02, ..., 52
folder_names = [
    "chu", "Ve", "Vao", "Tai", "Dau", "Cuoi", "Ke", "Truoc", "Dung", "Ngung",
    "Doc", "Tiep", "Lui", "Toi", "Tang", "To", "Giam", "Nho", "Lai", "Lap",
    "Nhanh", "Cham", "Luu", "Xoa", "Huy", "Chay", "Xong", "Dung", "Sai", "Giup",
    "Gio", "Ngay", "Tuoi", "Co", "Khong", "Muc", "Bai", "Mot", "Hai", "Ba",
    "Bon", "Nam", "Sau", "Bay", "Tam", "Chin", "A", "E", "I", "O",
    "U", "ThoDia"
]

# Tạo thư mục đích theo tên trong danh sách
for name in folder_names:
    os.makedirs(os.path.join(destination_dir, name), exist_ok=True)

# Duyệt qua các folder A, B, ..., T
for folder in sorted(os.listdir(source_dir)):
    folder_path = os.path.join(source_dir, folder)
    if os.path.isdir(folder_path):
        # Duyệt qua từng file trong folder
        for file in sorted(os.listdir(folder_path)):
            # Xác định số thứ tự của file (e.g., 'B01.wav' -> '01')
            file_index = int(file[1:3]) - 1  # Để phù hợp với index trong folder_names
            # Thư mục đích tương ứng
            target_folder = os.path.join(destination_dir, folder_names[file_index])
            # Đường dẫn nguồn và đích của file
            source_file = os.path.join(folder_path, file)
            destination_file = os.path.join(target_folder, file)
            # Di chuyển file
            shutil.move(source_file, destination_file)
            print(f"Moved {source_file} to {destination_file}")
```
